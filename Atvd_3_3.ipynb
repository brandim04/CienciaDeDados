```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler
```

```python
from ucimlrepo import fetch_ucirepo

# fetch dataset
adult = fetch_ucirepo(id=2)

# data (as pandas dataframes)
X = adult.data.features
y = adult.data.targets

# metadata
print(adult.metadata)

# variable information
print(adult.variables)
```

```python
print(X.head())
```

```python
X_processed = X.copy()

# Remover espaços extras e padronizar capitalização
for col in X_processed.select_dtypes(include=['object']).columns:
    X_processed[col] = X_processed[col].str.strip().str.title()  # Capitaliza a primeira letra de cada palavra

# Padronizar valores específicos, por exemplo países
country_mapping = {
    'United-States': 'United States',
    'United States': 'United States',
    'Self-Employed': 'Self Employed',  # Exemplo se houver variações
    # Adicione outros mapeamentos se necessário
}

X_processed['native-country'] = X_processed['native-country'].map(country_mapping).fillna(X_processed['native-country'])

# Converter valores "?" para NaN (já feito no read_csv, mas garantir)
X_processed.replace('?', np.nan, inplace=True)

# Verificar valores ausentes
print(X_processed.isnull().sum())

# Update X with the processed DataFrame
X = X_processed
```

```python
num_cols = X.select_dtypes(include=['int64', 'float64']).columns
print("Colunas numéricas:", list(num_cols))
```

```python
minmax_scaler = MinMaxScaler()
df_minmax = X.copy()
df_minmax[num_cols] = minmax_scaler.fit_transform(df_minmax[num_cols])
print("Exemplo Min-Max Scaling:")
print(df_minmax[num_cols].head())
```

```python
standard_scaler = StandardScaler()
df_standard = X.copy()
df_standard[num_cols] = standard_scaler.fit_transform(df_standard[num_cols])
print("Exemplo Standardization (Z-score):")
print(df_standard[num_cols].head())
```

```python
robust_scaler = RobustScaler()
df_robust = X.copy()
df_robust[num_cols] = robust_scaler.fit_transform(df_robust[num_cols])
print("Exemplo Robust Scaling:")
print(df_robust[num_cols].head())
```

```python
# Estatísticas básicas para comparação
print("Original:")
print(X[num_cols].describe())

print("\nMin-Max Scaling:")
print(df_minmax[num_cols].describe())

print("\nZ-score Standardization:")
print(df_standard[num_cols].describe())

print("\nRobust Scaling:")
print(df_robust[num_cols].describe())
```
